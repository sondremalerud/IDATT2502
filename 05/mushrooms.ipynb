```python
# Oppgave 1, 2, 3

import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np

data = pd.read_csv('./data/mushroom/expanded')
data
```

```python
data.describe()
```

```python
print("EDIBLE: " + str(data['poisonous'][data['poisonous'] == 'EDIBLE'].count()))
print("POISONOUS: " + str(data['poisonous'][data['poisonous'] == 'POISONOUS'].count()))
```

```python
# Oppgave 4

habitats = data.groupby(['habitat', 'poisonous']).size().unstack()
habitats.plot(kind='bar', stacked=True)
plt.ylabel('Count')
plt.xlabel('Habitat')
plt.show()
```

```python
# Oppgave 5
dummy = pd.get_dummies(data)
dummy
```

```python
# Oppgave 6

# Plot document-term matrix, blue indicates non-zero value
plt.spy(dummy, markersize=1)
fig = plt.gcf()
fig.set_size_inches(60,400)
plt.plot()
plt.show()
```
